```python
from keras.preprocessing.image import ImageDataGenerator
import os
from matplotlib import pyplot as plt
```

```python
#Definimos el data generator. Aqui se pueden aplicar cualquier transformacion a las imagenes
datagen = ImageDataGenerator()
```

```python
#declarar directorio de entrenamiento con carpetas ordenadas por clases
train_dir = os.getcwd() + "/reorganized/"
#usar flow_from_directory
train_data_keras = datagen.flow_from_directory(
    directory=train_dir,
    class_mode='categorical',
    batch_size=16, #16 imagenes a la vez
    target_size=(32,32)) #dimension definida para todas las imagenes
    
```

```python
#Revisar las imagenes de un batch
x, y = next(train_data_keras)

#Ver cada imagen

for i in range (0,15):
    image = x[i].astype(int)
    plt.imshow(image)
    plt.show()
```
